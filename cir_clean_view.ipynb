```python
import pandas as pd 
import numpy as np

```

```python
#sheets = ["Y10", "Y11", "Y12", "Y13", "Y14", "Y15", "Y16", "Y17", "Y18"]
sheets = ["Y11", "Y12"]
master_df = pd.DataFrame(columns=['Segment', 'Category', 'Region', 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'AVG'])
for sheet in sheets:
    Y_df = pd.read_excel("Industry_42_CIR_practice.xlsx", sheet)
    headers = Y_df.iloc[0]
    intial_columns_df = pd.DataFrame(Y_df.values[1:], columns=headers).iloc[0:52, 0:3]
    intial_columns_df.columns.values[0] = "Segment"
    intial_columns_df.columns.values[1] = "Category"
    intial_columns_df.columns.values[2] = "Region"
    intial_columns_df["Year"] = sheet

    new_df  = pd.DataFrame(Y_df.values[1:], columns=headers)
    df0 = new_df.iloc[0:52, 3:16]
    df1 = new_df.iloc[0:52, 17:30]
    df2 = new_df.iloc[0:52, 31:44]
    df3 = new_df.iloc[0:52, 45:58]

    df0 = pd.concat([intial_columns_df, df0], axis=1)
    df0["Region"] = "NorthA"
    df0["Segment"][0:19] = "INTERNET"
    df0["Segment"][20:42] = "WHOLESALE"
    df0["Segment"][43:51] = "PRIVATELABEL"

    df1 = pd.concat([intial_columns_df, df1], axis=1)
    df1["Region"] = "EP"
    df1["Segment"][0:19] = "INTERNET"
    df1["Segment"][20:42] = "WHOLESALE"
    df1["Segment"][43:51] = "PRIVATELABEL"

    df2 = pd.concat([intial_columns_df, df2], axis=1)
    df2["Region"] = "AP"
    df2["Segment"][0:19] = "INTERNET"
    df2["Segment"][20:42] = "WHOLESALE"
    df2["Segment"][43:51] = "PRIVATELABEL"

    df3 = pd.concat([intial_columns_df, df3], axis=1)
    df3["Region"] = "LA"
    df3["Segment"][0:19] = "INTERNET"
    df3["Segment"][20:42] = "WHOLESALE"
    df3["Segment"][43:51] = "PRIVATELABEL"

    master_df = pd.concat([master_df, df0, df1, df2, df3])

master_df


```

```python
master_df = master_df.replace(r'^\s*$', np.nan, regex=True)
master_df = master_df.dropna(axis=0)
```
